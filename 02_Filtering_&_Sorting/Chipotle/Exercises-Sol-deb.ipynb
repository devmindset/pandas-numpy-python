### Step 1. Import the necessary libraries

```python
import numpy as np
```

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

### Step 3. Assign it to a variable called chipo.

```python
chipo=pd.read_csv(url, sep='\t')
```

```python
chipo
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].replace( '[\$]','', regex=True ).apply(np.float64)
```

```python
len(chipo[chipo['item_price'] >10].groupby('item_name').count())
```

```python
chipo[chipo['item_price'] >10].item_name.nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo[['item_price','item_name']].head(5)
```

### Step 6. Sort by the name of the item

```python
chipo[['item_price','item_name']].sort_values(['item_name'], ascending=[False]).head(5)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
maxprice = chipo[['item_price','item_name']].max()['item_price']
```

```python
maxprice
```

```python
chipo[['item_price','item_name','quantity']].sort_values(['item_price'], ascending=[False]).head(5)
```

```python
chipo[chipo['item_price'] == maxprice].quantity.sum()
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
len(chipo[chipo['item_name'] == "Veggie Salad Bowl"])
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
len(chipo[(chipo['item_name'] == "Canned Soda") & (chipo['quantity'] > 1)])
```
