### Step 13. Turn the item price into a float

```python
chipo2 = pd.read_csv(url, sep='\t')
```

```python
chipo2.info()
```

```python
chipo2['item_price'].replace( '[\$]','', regex=True ).astype(float).head(10)
```

```python
chipo2.info()
```

```python
chipo['item_price'] = chipo['item_price'].replace( '[\$]','', regex=True ).apply(np.float64)
```

```python
chipo.head(10)
```

### Step 14. How much was the revenue for the period in the dataset?

```python
chipo.info()
```

```python
chipo['revenue'] = chipo.quantity*chipo.item_price
```

```python
chipo['revenue'].sum()
```

### Step 15. How many orders were made in the period?

### Step 16. What is the average amount per order?

```python
chipo.groupby('order_id').sum().head()
```

```python
chipo.groupby('order_id').sum()['revenue'].mean()
```

```python
chipoGroupbyOderIdSum = chipo.groupby('order_id').sum() 
```

```python
(chipoGroupbyOderIdSum['revenue']/chipoGroupbyOderIdSum['quantity']).head()
```
